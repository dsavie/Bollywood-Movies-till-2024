```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Correct file path for the uploaded file
file_path = '/content/.config/logs/2024.12.18/Bollywood Movie List (1920-2024).csv'

# Load the dataset
df = pd.read_csv(file_path)

# Display the first few rows
print(df.head())

```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df.fillna(method='ffill')  # Forward-fill missing values
```

```python
df = df.drop_duplicates()
```

```python
df['Year'] = df['Movie ID'].fillna(0)  # Replace NaN with 0
```

```python
movies_by_year = df['Year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
movies_by_year.plot(kind='bar')
plt.title('Number of Movies by Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numerical_df.corr()

# Plot the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Check column data types
print(df.dtypes)

# Check for non-numeric values in a specific column
print(df['Movie ID'].unique())  # Replace 'ColumnName' with the column name
```

```python
movies_trend = df.groupby('Year').size()
plt.figure(figsize=(12, 6))
plt.plot(movies_trend, marker='o')
plt.title('Movies Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid()
plt.show()
```
